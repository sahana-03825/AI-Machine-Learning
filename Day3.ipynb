- Load and Explore The Dataset

```python
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/Naresh312143/mahalingpur-python-ml/refs/heads/main/Salary_Data.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

- Feature Scaling(Very Important For SVR)

```python
from sklearn.preprocessing import StandardScaler

X=df[['Years of Experience']].values
y=df[['Salary']].values

sc_X=StandardScaler()
sc_y=StandardScaler()

X_scaled=sc_X.fit_transform(X)
y_scaled=sc_y.fit_transform(y)
```

- Flatten for SVR

```python
y_scaled
```

```python
y_scaled =y_scaled.ravel()
```

```python
y_scaled
```

- Train SVR Model(RBF Kernel)

```python
from sklearn.svm import SVR

svr=SVR(kernel='rbf',C=100,epsilon=0.1)
svr.fit(X_scaled,y_scaled)
```

- Predictions And Inverse Scaling

```python
y_pred_scaled=svr.predict(X_scaled)
y_pred=sc_y.inverse_transform(y_pred_scaled.reshape(-1,1))
```

```python
y_pred_scaled
```

```python
y_pred
```

- Visualization:Actual vs Predicted

```python
import matplotlib.pyplot as plt

plt.scatter(X,y,color='red',label='Actual')
plt.plot(X,y_pred,color='blue',label='SVR Prediction')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title("SVR-Salary Prediction")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Create a fine gird between min and max YearsExperience
X_grid=np.arange(min(X.ravel()),max(X.ravel()),0.1)
X_grid=X_grid.reshape(-1,1)

#Scale X_gird just like training data
X_gird_scaled=sc_X.transform(X_grid)

#Predict on the gird and inverse transform the results
y_grid_scaled=svr.predict(X_gird_scaled)
y_grid=sc_y.inverse_transform(y_grid_scaled.reshape(-1,1))

#Plot
plt.figure(figsize=(8,5))
plt.scatter(X,y,color='red',label='Actual')
plt.plot(X_grid,y_grid,color='blue',label='SVR Smooth Curve')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title("SVR (RBF Kernal)-Smooth Prediction Curve")
plt.grid(True)
plt.legend()
plt.show()
```

- Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np

mse=mean_squared_error(y,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y,y_pred)

print("MSE:",mse)
print(f"RMSE:",rmse)
print("r2 Score:",r2)

```

- Compare Kernals Using Code

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

#1.Load and scale the data
X=df[['Years of Experience']].values
y=df[['Salary']].values

sc_X=StandardScaler()
sc_y=StandardScaler()

X_scaled=sc_X.fit_transform(X)
y_scaled=sc_y.fit_transform(y).ravel()

#2.Define kernels
kernels=['linear','poly','rbf']

#3.Train and evaluate each kernel
for kernel in kernels:
  if kernel=='poly':
    model=SVR(kernel=kernel,degree=2)
  else:
    model=SVR(kernel=kernel)

  model.fit(X_scaled,y_scaled)

    #Predict for learning data
  y_pred_scaled=model.predict(X_scaled)
  y_pred=sc_y.inverse_transform(y_pred_scaled.reshape(-1,1))

    #Evaluation
  score=r2_score(y,y_pred)
  print(f"Kernel:{kernel.upper()} | R_squared Score:{score:.4f}")

    #Plot(optional)
  plt.scatter(X,y,color='red',label='Actual Data')
  plt.plot(X,y_pred,label=f'{kernel.upper()}Kernel',linewidth=2)
  plt.xlabel('Years of Experience')
  plt.ylabel('Salary')
  plt.title(f'SVR using {kernel.upper()}kernal')
  plt.grid(True)
  plt.legend()
  plt.show()
```
